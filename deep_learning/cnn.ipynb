```python
from sklearn.datasets import load_sample_image
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# loading images
china = load_sample_image('china.jpg') / 255
flower = load_sample_image('flower.jpg') / 255
imgs = np.array([china, flower])
batch_size, h, w, ch = imgs.shape
print(china.shape)
print(imgs.shape)
```

```python
# creating 2 filters
filters = np.zeros(shape=(7,7, ch, 2), dtype=np.float32)
filters[:, 3, :, 0] = 1 # vertical line
filters[3, :, :, 1] = 1 # horizontal line
```

```python
op = tf.nn.conv2d(imgs, filters, strides=1, padding='SAME')
```

```python
plt.imshow(op[0, :, :, 1], cmap='gray')
plt.show()
```
