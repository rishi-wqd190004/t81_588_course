2. Use *quantile* values to get rows

## Target Encoding
- Generally used on categorical columns only and for regression based problems
- Calculate avergage target value for that category, then to encode we substitute the percent corresponding to the category that the categorical value has.

# Another example

```python
data = pd.read_csv("https://www.irs.gov/pub/irs-soi/16zpallagi.csv")
```

```python
data.head(2)
```

```python
data = data.loc[
    (data['zipcode'] !=0) & (data['zipcode'] != 99999),
    ['STATE', "zipcode", 'agi_stub', 'N1'],
]
data.head(2)
```

```python
data[data['zipcode'] == 63017]
```

```python
median_n1 = {
    1: 12500,
    2: 37500,
    3: 62500,
    4: 87500,
    5: 112500,
    6: 212500
}
data['median_agi_stub'] = data['agi_stub'].map(median_n1)
```

```python
data[data['zipcode'] == 63017]
```

```python
grps = data.groupby(by='zipcode')
```
